# Outliers in Machine Learning

Outlier detection is a crucial aspect of machine learning, focusing on identifying data points that deviate significantly from the majority of the dataset. These anomalies can indicate errors, rare events, or novel insights, depending on the application.

---

## Anomaly Detection Applications
- **Fraud Detection:** Identifying unusual transactions in financial systems.
- **Network Security:** Detecting suspicious activities or breaches.
- **Healthcare:** Spotting abnormal patient vitals or medical records.
- **Manufacturing:** Monitoring equipment for faults and defects.
- **Finance:** Identifying irregular trading patterns.
- **Sensor Networks:** Detecting faulty sensor readings.

## Types of Anomalies
1. **Point Anomalies:** Single data points that are significantly different from the rest.
2. **Contextual Anomalies:** Anomalies contextual to a specific situation (e.g., low temperature in summer).
3. **Collective Anomalies:** A group of data points that collectively exhibit abnormal behavior (e.g., high volume of requests).
4. **Temporal Anomaly Patterns:** Irregular patterns in time series data.
5. **Spatial Data Anomalies:** Unusual patterns in spatial data.

### Detailed Anomaly Types
- **Point Anomalies:**  
  Example: A transaction amount that is unusually high compared to typical transactions.
  
- **Contextual Anomalies:**  
  Example: A low temperature reading during summer months.
  
- **Collective Anomalies:**  
  Example: A sudden spike in web traffic indicating a potential Distributed Denial of Service (DDoS) attack.
  
- **Temporal Anomaly Patterns:**  
  Example: Unexpected changes in stock prices over time.
  
- **Spatial Data Anomalies:**  
  Example: Unusual geographic distribution of events in spatial data.

## Detection Methods
- **Z-Score:**  
  Measures how many standard deviations a data point is from the mean.  
  *Note: Popular but less effective for complex datasets.*

- **Statistical Hypothesis Testing:**  
  Determines anomalies based on statistical significance.

- **Density-Based Methods:**  
  Identify regions with low data density, effective for various datasets.  
  *Example: Local Outlier Factor (LOF).*

- **Probabilistic Models:**  
  Assess the probability of data points belonging to a distribution.

- **Regression Analysis:**  
  Detects points with significant deviations after applying regression models.

- **Principal Component Analysis (PCA):**  
  Identifies anomalies in transformed feature space by reducing dimensionality.

- **Time Series Decomposition:**  
  Analyzes trend, seasonality, and residual components to spot anomalies.

- **Exponential Smoothing Models:**  
  Forecasts and detects anomalies based on smoothed trends.

## Popular Models
### Density-Based Models
1. **K-Nearest Neighbors (KNN):**  
   Measures the distance to the k-nearest neighbors to identify outliers.

2. **Local Outlier Factor (LOF):**  
   Evaluates the local density deviation of a data point compared to its neighbors.

### Support Vector Machines
3. **One-Class SVM:**  
   Trains on the normal class and identifies novelties or outliers.

### Tree-Based Models
4. **Isolation Forest:**  
   Isolates anomalies by recursively partitioning the data.  
   *Note: One of the most effective methods for outlier detection.*

### Non-Density-Based Models
5. **Clustering Algorithms:**  
   Identify group-based anomalies by analyzing cluster memberships.

6. **Neural Networks:**  
   Powerful for complex data but require significant resources.

7. **Gaussian Mixture Models (GMM):**  
   Models data as a mixture of multiple Gaussian distributions.

8. **Bayesian Networks:**  
   Utilize probabilistic graphical models to identify anomalies.

## Deep Learning Approaches
- **Recurrent Neural Networks (RNNs):**  
  Suitable for sequential or time series data.

- **Generative Adversarial Networks (GANs):**  
  Generate and detect anomalies by training generator and discriminator networks.

- **Convolutional Neural Networks (CNNs):**  
  Effective for spatial data anomalies, such as image data.

- **Variational Autoencoders (VAE):**

- **Deep Belief Networks (DBN):**

- **Attention Mechanisms:**

- **Hybrid Models:**  
  Combine multiple approaches for improved detection performance.

## Evaluation Metrics
- **F1-Score:**  
  Balances precision and recall; the most important metric for outlier detection.

- **Precision & Recall:**  
  - **Precision:** The accuracy of the positive predictions.
  - **Recall:** The ability to find all relevant positive instances.

- **AUC & ROC:**  
  Evaluate model performance across different thresholds.

- **Mean Squared Error (MSE):**  
  Measures prediction errors, useful in regression-based anomaly detection.

- **Confusion Matrix:**  
  Provides detailed classification performance, showing true vs. predicted labels.

*Note:* While F1-Score, Precision, and Recall are primarily for supervised classification, outlier detection is often a semi-supervised approach with approximately 1%-3% of the dataset labeled as anomalies.

### How Isolation Forest Works (Simplified)

1. **Goal**: Detect anomalies (outliers) in data.
2. **Key Idea**: Anomalies are rare and different, so they’re easier to isolate.

---

### Steps:
1. **Build Isolation Trees (iTrees)**:
   - Randomly pick a feature and a split value.
   - Split the data into two branches.
   - Repeat until data points are isolated or a depth limit is reached.

2. **Measure Path Length**:
   - Count how many splits it takes to isolate a point.
   - Anomalies need fewer splits (shorter paths).
   - **Normal points need more splits (longer paths)** because they are more densely distributed.

3. **Calculate Anomaly Score**:
   - Average the path lengths across all trees.
   - Normalize the score using the formula:
     \[
     s(x, n) = 2^{-\frac{E(h(x))}{c(n)}}
     \]
     - \(E(h(x))\): Average path length for point \(x\).
     - \(c(n)\): Normalization factor based on dataset size \(n\).
   - **Score ranges from 0 to 1**:
     - **Close to 1**: Anomaly.
     - **Close to 0**: Normal.
     - **Around 0.5**: Neither clearly normal nor anomalous.

4. **Ensemble of Trees**:
   - Build multiple iTrees (a "forest") to improve accuracy and robustness.
   - Each tree is trained on a random subset of the data.
   - The final anomaly score is the average across all trees.

---

### Why It’s Great:
- Fast and works well with big, high-dimensional data.
- Doesn’t need labeled data (unsupervised).
- Ensemble approach reduces variance and improves reliability.

## How LOF Works

The **Local Outlier Factor (LOF)** algorithm identifies anomalies by comparing the local density of each data point to that of its neighbors. Here's a simplified breakdown of how LOF operates:

### 1. **Select the Number of Neighbors (k)**
- **k** determines how many nearby points are considered when assessing the density around a data point.
- A typical starting value is between 10 and 20, but it can be adjusted based on the dataset's characteristics.

### 2. **Determine k-Nearest Neighbors**
- For each data point, identify its **k** closest neighbors using a distance metric (e.g., Euclidean distance).
- This set of neighbors forms the local context for density comparison.

### 3. **Calculate Reachability Distance**
- **Reachability Distance** measures how far a point is from its neighbors relative to their own distances.
- For a data point **A** and its neighbor **B**, the reachability distance is the greater of:
  - The actual distance between **A** and **B**.
  - The distance from **B** to its k-th nearest neighbor.
- This adjustment ensures that sparse regions influence the density measurement appropriately.

why do we accept GREATER one? If point A is near B, but B is in a sparse zone (neighbors far away), then A should also be considered in a sparse zone.



### 4. **Compute Local Reachability Density (LRD)**
- **LRD** quantifies the density around a data point based on the reachability distances to its neighbors.
- It is calculated as the inverse of the average reachability distance from the data point to its k neighbors.
  
  **LRD(A) = 1 / ( (ReachDist(A, B1) + ReachDist(A, B2) + ... + ReachDist(A, Bk)) / k )**

### 5. **Calculate the LOF Score**
- The **LOF score** for a data point compares its LRD to the average LRD of its neighbors.
  
  **LOF(A) = (LRD(B1) + LRD(B2) + ... + LRD(Bk)) / (k * LRD(A))**
  
- **Interpretation of LOF Scores**:
  - **LOF ≈ 1**: The data point has a density similar to its neighbors (normal point).
  - **LOF > 1**: The data point is less dense than its neighbors (potential outlier).
  - **Higher LOF Values**: Indicate a higher likelihood of being an outlier.

### 6. **Identify Outliers**
- After computing LOF scores for all data points, those with scores significantly greater than 1 are flagged as outliers.
- The threshold for determining outliers can be set based on the desired sensitivity or by analyzing the distribution of LOF scores.

### **Example Workflow**

1. **Choose k = 5**.
2. **For each data point**, find its 5 nearest neighbors.
3. **Compute the reachability distance** for each neighbor.
4. **Calculate the LRD** for each data point.
5. **Determine the LOF score** by comparing each point's LRD to its neighbors'.
6. **Flag points with LOF > 1.5** as outliers.

### **Visual Illustration**

Imagine a dataset where most points are densely packed in clusters, but some points lie isolated or in sparse regions:

- **Normal Points**: Located within dense clusters, having similar LOF scores close to 1.
- **Outliers**: Situated in less dense areas or far from clusters, exhibiting higher LOF scores.

By assessing the local density, LOF effectively distinguishes outliers that deviate from their immediate surroundings, even in complex and varied datasets.

```python
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame with 'value' and 'is_outlier' columns
# Ensure that 'is_outlier' is binary: 1 for outliers, 0 for inliers

# Split the data
train = df.iloc[:7001]['value'].values.reshape(-1, 1)  # Reshape to 2D
test = df.iloc[7000:]['value'].values.reshape(-1, 1)   # Reshape to 2D

# Initialize StandardScaler
scaler = StandardScaler()

# Fit on training data and transform both training and test data
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)

# Initialize One-Class SVM
# 'nu' is an upper bound on the fraction of training errors (outliers)
nu_value = (3 / train.shape[0]) * 2  # Adjust as needed
ocsvm = OneClassSVM(kernel='rbf', gamma='scale', nu=nu_value)

# Fit the model
ocsvm.fit(train_scaled)

# Predict on test data
predictions = ocsvm.predict(test_scaled)
print(np.unique(predictions))  # Outputs: [-1  1]

# Convert labels from {-1, 1} to {1, 0}
# Here, -1 (outlier) becomes 1 and 1 (inlier) becomes 0
labels = np.where(predictions == -1, 1, 0)

# Compute confusion matrix
cm = confusion_matrix(df.iloc[7000:]['is_outlier'], labels)
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Inlier', 'Outlier'])
cmd.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - One-Class SVM with Scaling")
plt.show()

# Display classification report
print(classification_report(df.iloc[7000:]['is_outlier'], labels, target_names=['Inlier', 'Outlier']))
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector
from sklearn.preprocessing import MinMaxScaler

# -------------------------------
# 1. Data Preparation
# -------------------------------

# Parameters
sequence_length = 10  # Number of time steps per sequence
train_size = 7001     # Number of samples for training

# Normalize the data
scaler = MinMaxScaler()
df['value_scaled'] = scaler.fit_transform(df[['value']])

# Function to create sequences
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        seq = data[i:i + seq_length]
        sequences.append(seq)
    return np.array(sequences)

# Create training and testing sequences
train_data = create_sequences(df.iloc[:train_size]['value_scaled'].values, sequence_length)
test_data = create_sequences(df.iloc[train_size - sequence_length:]['value_scaled'].values, sequence_length)

# Reshape to add feature dimension (required for LSTM)
# New shape: (samples, sequence_length, features)
train_data = train_data.reshape((train_data.shape[0], train_data.shape[1], 1))
test_data = test_data.reshape((test_data.shape[0], test_data.shape[1], 1))

# -------------------------------
# 2. Building the LSTM Autoencoder
# -------------------------------

# Define the Autoencoder architecture
input_layer = Input(shape=(sequence_length, 1))
encoded = LSTM(64, activation='relu')(input_layer)
decoded = RepeatVector(sequence_length)(encoded)
decoded = LSTM(64, activation='relu', return_sequences=True)(decoded)
decoded = LSTM(1, activation='sigmoid', return_sequences=True)(decoded)

# Compile the model
autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# -------------------------------
# 3. Training the Model
# -------------------------------

# Train the Autoencoder
autoencoder.fit(
    train_data,
    train_data,
    epochs=20,
    batch_size=32,
    validation_split=0.1,
    shuffle=True
)

# -------------------------------
# 4. Anomaly Detection
# -------------------------------

# Predict on test data
X_pred = autoencoder.predict(test_data)
# Calculate reconstruction error (MSE) for each sequence
mse = np.mean(np.power(test_data - X_pred, 2), axis=(1, 2))

# Set threshold for anomaly detection (e.g., 95th percentile)
threshold = np.percentile(mse, 95)
print(f"Threshold for anomaly detection: {threshold}")

# Label sequences as outliers (1) or inliers (0)
labels = (mse > threshold).astype(int)

# -------------------------------
# 5. Evaluation
# -------------------------------

# Align true labels with the test sequences
# Since sequences are created by sliding window, adjust true_labels accordingly
true_labels = df.iloc[train_size:]['is_outlier'].values[:len(labels)]

# Compute confusion matrix
cm = confusion_matrix(true_labels, labels)
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Inlier', 'Outlier'])
cmd.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - LSTM Autoencoder")
plt.show()

# Print classification report
print(classification_report(true_labels, labels, target_names=['Inlier', 'Outlier']))
```

# Data Scaling in Machine Learning Models

## Importance of Scaling the Dataset

**Data scaling** is a crucial preprocessing step in machine learning that involves adjusting the range of feature values. Scaling ensures that each feature contributes equally to the result, preventing biases due to differing scales.

### **Why Scale the Dataset?**
- **Uniform Contribution:** Ensures all features contribute proportionately to distance-based algorithms.
- **Improved Convergence:** Helps optimization algorithms converge faster by providing a balanced feature space.
- **Enhanced Performance:** Prevents features with larger scales from dominating the learning process.

Common scaling techniques include:
- **StandardScaler:** Centers the data by removing the mean and scales it to unit variance.
- **MinMaxScaler:** Scales features to a fixed range, typically [0, 1].

## Scaling and Different Outlier Detection Methods

### **1. Isolation Forest**
- **Why Scaling Often Unnecessary:**
  - **Tree-Based Method:** Isolation Forest is a tree-based algorithm that splits data based on feature values without relying on distance metrics.
  - **Insensitive to Feature Scales:** Since it uses random splits, the scale of features has minimal impact on its performance.
  
- **When to Consider Scaling:**
  - **Multiple Features with Different Scales:** If your dataset has features with vastly different scales, scaling might still benefit the model by providing more balanced splits.

### **2. Local Outlier Factor (LOF)**
- **Why Scaling is Essential:**
  - **Distance-Based:** LOF relies on distance metrics (e.g., Euclidean distance) to assess the local density around each data point.
  - **Feature Scale Impact:** Features with larger scales can disproportionately influence distance calculations, skewing outlier detection.
  
- **Recommendation:** Always scale your data when using LOF to ensure accurate density comparisons.

### **3. Support Vector Machine (SVM) - One-Class SVM**
- **Why Scaling is Crucial:**
  - **Kernel-Based Method:** SVMs use kernel functions that compute distances between data points in transformed feature spaces.
  - **Sensitivity to Feature Magnitude:** Uneven feature scales can distort the kernel calculations, leading to suboptimal decision boundaries.
  
- **Recommendation:** Scale your data to standardize feature contributions and improve model performance.

### **4. Long Short-Term Memory (LSTM) Networks**
- **Why Scaling Benefits LSTM:**
  - **Neural Network Training:** LSTMs, like other neural networks, perform better when input features are scaled, as it stabilizes and accelerates the training process.
  - **Gradient Descent Efficiency:** Properly scaled data ensures more efficient gradient updates, preventing issues like vanishing or exploding gradients.
  
- **Recommendation:** Normalize or standardize your data before feeding it into LSTM models to enhance learning efficiency and model accuracy.

## Conclusion

Scaling is a fundamental preprocessing step for many machine learning algorithms, especially those sensitive to feature magnitudes and distance calculations. While tree-based methods like **Isolation Forest** are less impacted by feature scales, algorithms such as **LOF**, **One-Class SVM**, and **LSTM** significantly benefit from scaled data to ensure balanced feature contributions and optimal performance. Always consider the nature of your algorithm and dataset when deciding to apply scaling.
