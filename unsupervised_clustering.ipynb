## Finding Optimal K for K-Means

1. **Elbow Method:**  
   - Plot the Within-Cluster-Sum-of-Squares (WCSS) vs. K.  
   - If every point is its own cluster (K = number of points), WCSS = 0.  
   - The "elbow" is where WCSS stops decreasing significantly, indicating the optimal number of clusters.  
   - **Choosing K after the elbow may lead to overfitting**, as it creates more clusters than necessary to describe the data meaningfully.

2. **Silhouette Score:**  
   Measure how well-separated clusters are; higher scores indicate better-defined clusters.

3. **Gap Statistic:**  
   Compare the clustering result with random data to find the optimal K.

4. **Domain Knowledge:**  
   Use prior knowledge about the data to guide the choice of K.


## Mini-Batch K-Means

- **What is it?**  
  A faster, memory-efficient variant of K-Means that processes data in small batches instead of the full dataset.

- **Advantages:**  
  - Scales well to large datasets.  
  - Faster convergence compared to standard K-Means.  
  - Reduces computational cost by updating centroids with subsets of data.

- **How it Works:**  
  1. Randomly select a small batch of data points.  
  2. Update centroids based on the batch.  
  3. Repeat until convergence.

- **Limitations:**  
  - Slightly less accurate than standard K-Means.  
  - Requires careful tuning of batch size.

- **Use Case:**  
  Suitable for clustering large datasets where full K-Means is computationally prohibitive.


# Clustering Metrics

## **Metrics Requiring Ground Truth Labels**
1. **Adjusted Rand Index (ARI):**  
   Measures the similarity between predicted and true clusters, corrected for chance.

2. **Normalized Mutual Information (NMI):**  
   Quantifies the amount of information shared between true and predicted clusters.

3. **Fowlkes-Mallows Index (FMI):**  
   Evaluates the precision and recall of predicted clusters compared to true clusters.

---

## **Metrics Without Ground Truth Labels**
1. **Silhouette Score:**  
   Measures how well clusters are separated based on intra-cluster and inter-cluster distances.

2. **Calinski-Harabasz Index:**  
   Evaluates the ratio of between-cluster dispersion to within-cluster dispersion.

3. **Davies-Bouldin Index:**  
   Measures the average similarity ratio of each cluster to its most similar cluster.

---

### **Key Point**
- **Metrics Requiring Ground Truth Labels:** ARI, NMI, FMI.  
- **Metrics Without Ground Truth Labels:** Silhouette, Calinski-Harabasz, Davies-Bouldin.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans , DBSCAN , MiniBatchKMeans , AgglomerativeClustering
from sklearn.metrics import silhouette_score , adjusted_rand_score , normalized_mutual_info_score , make_scorer
from sklearn.preprocessing import StandardScaler , MinMaxScaler
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV
```

```python
X,y = make_blobs(n_samples=500 , centers=4 , random_state=42 , n_features=2 ,cluster_std=1.5)
X
```

```python
plt.scatter(X[:,0] , X[:,1] , c=y , cmap='coolwarm' ,alpha=.6)
plt.title('Synthetic dataset')
```

```python
inertia = []
ks=range(2,10)
for k in ks:
  km=KMeans(n_clusters=k , random_state=13 )
  pred_labels = km.fit_predict(X)
  print(f'for k={k}')
  print(f'KMeans inertia={km.inertia_}')
  print(f'ARI={adjusted_rand_score(y , pred_labels)}')
  print(f'Sillhouette Index={silhouette_score(X , pred_labels)}')
  print('')
  inertia.append(km.inertia_)
```

```python

url = 'https://cs.joensuu.fi/sipu/datasets/jain.txt'

# Define column names based on your data structure
column_names = ['Feature1', 'Feature2', 'Label']

# Read the CSV with correct separator and column names
df = pd.read_csv(url, sep='\t', header=None, names=column_names)

ssc = MinMaxScaler()
scaled_data = ssc.fit_transform(df[['Feature1', 'Feature2']])
df = pd.concat([pd.DataFrame(scaled_data,columns=['Feature1', 'Feature2']) , df['Label']] ,axis=1)
print(df.sample(10))
plt.scatter(df['Feature1'] , df['Feature2'] , c=df['Label'])

```

```python

# Initialize clustering algorithms with chosen parameters
dbscan = DBSCAN(eps=0.081, min_samples=1)  # Adjust eps and min_samples as needed
kmeans = KMeans(n_clusters=2, random_state=42)
agg_clustering = AgglomerativeClustering(n_clusters=2)

# Fit and predict cluster labels
labels_dbscan = dbscan.fit_predict(df[['Feature1', 'Feature2']])
labels_kmeans = kmeans.fit_predict(df[['Feature1', 'Feature2']])
labels_agg = agg_clustering.fit_predict(df[['Feature1', 'Feature2']])

# -------------------------------
# 4. Evaluating Clustering Performance
# -------------------------------

# Calculate Silhouette Scores to evaluate clustering performance
# Note: Silhouette Score requires at least 2 clusters and less than number of samples
try:
    silhouette_dbscan = silhouette_score(df[['Feature1', 'Feature2']], labels_dbscan)
except:
    silhouette_dbscan = "N/A (Only one cluster or all points are noise)"

try:
    silhouette_kmeans = silhouette_score(df[['Feature1', 'Feature2']], labels_kmeans)
except:
    silhouette_kmeans = "N/A"

try:
    silhouette_agg = silhouette_score(df[['Feature1', 'Feature2']], labels_agg)
except:
    silhouette_agg = "N/A"

print("\nSilhouette Scores:")
print(f"DBSCAN: {silhouette_dbscan}")
print(f"K-Means: {silhouette_kmeans}")
print(f"Agglomerative Clustering: {silhouette_agg}")

# -------------------------------
# 5. Visualization Function
# -------------------------------

def plot_clusters(df, labels, title, ax):
    """
    Plots the clustering results on a given Axes object.

    Parameters:
    - df: pandas DataFrame containing the data.
    - labels: Cluster labels for each data point.
    - title: Title of the plot.
    - ax: Matplotlib Axes object to plot on.
    """
    unique_labels = set(labels)
    colors = plt.cm.viridis(np.linspace(0, 1, len(unique_labels)))

    for label, color in zip(unique_labels, colors):
        if label == -1:
            # Typically, DBSCAN labels noise as -1
            color = 'k'
            label_name = 'Noise'
        else:
            label_name = f'Cluster {label}'

        ax.scatter(
            df.loc[labels == label, 'Feature1'],
            df.loc[labels == label, 'Feature2'],
            c=[color],
            label=label_name,
            edgecolor='k',
            s=50
        )

    ax.set_title(title)
    ax.set_xlabel('Feature1')
    ax.set_ylabel('Feature2')
    ax.legend()

# -------------------------------
# 6. Plotting Clustering Results
# -------------------------------

# Create subplots for each clustering result
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot DBSCAN
plot_clusters(df[['Feature1', 'Feature2']], labels_dbscan, 'DBSCAN Clustering', axes[0])

# Plot K-Means
plot_clusters(df[['Feature1', 'Feature2']], labels_kmeans, 'K-Means Clustering', axes[1])

# Plot Agglomerative Clustering
plot_clusters(df[['Feature1', 'Feature2']], labels_agg, 'Agglomerative Clustering', axes[2])

plt.tight_layout()
plt.show()
```
