# Dimensionality Reduction

Dimensionality reduction is a vital technique in data preprocessing and machine learning that reduces the number of input features in a dataset while preserving essential information. This simplification enhances model performance, interpretability, and computational efficiency.

## Why Dimensionality Reduction?

- **Reduce Features Without Losing Information:** Eliminates redundant or less informative features.
- **Improve Performance:** Speeds up training and enhances algorithm efficiency.
- **Reduce Overfitting:**
  - **Causes:** More features increase model complexity and risk of memorizing noise.
  - **Effects:** Poor generalization to unseen data.
- **Facilitate Visualization:** Enables plotting high-dimensional data in 2D or 3D.
- **Enhance Feature Extraction:**
  - **Selection:** Choose the most relevant existing features.
  - **Creation:** Transform or combine original features into new ones.

## Types of Dimensionality Reduction

### 1. Feature Selection
Selects a subset of original features based on specific criteria.
- **Advantages:** Maintains interpretability and reduces training complexity.
- **Techniques:** Filter methods, wrapper methods, embedded methods.

### 2. Feature Extraction
Transforms data into a lower-dimensional space by creating new features.
- **Advantages:** Captures complex relationships and often results in uncorrelated features.
- **Techniques:** PCA, LDA, t-SNE, UMAP, Autoencoders, NMF.

## Methods of Dimensionality Reduction

### 1. Principal Component Analysis (PCA)
**Linear** technique that identifies directions of maximum variance.
- **Key Features:** Uncorrelated principal components, information distribution across components.
- **Time Consumption:** \(O(n^3)\) complexity; less efficient for large datasets.
- **Use Cases:** Noise reduction, feature visualization, preprocessing.

### 2. Linear Discriminant Analysis (LDA)
**Linear, supervised** technique focusing on class separability.
- **Key Features:** Maximizes between-class variance vs. within-class variance.
- **Time Consumption:** Similar to PCA; requires labeled data.
- **Use Cases:** Classification, pattern recognition.

### 3. t-Distributed Stochastic Neighbor Embedding (t-SNE)
**Non-linear** technique for high-quality visualizations.
- **Key Features:** Preserves local structures, creates interpretable clusters.
- **Time Consumption:** \(O(n^2)\) complexity; slow for large datasets.time complexity of t-SNE is lower than PCA but runtime for LDA usually takes so long because time complexity is high and this is a non-linear apporach!
- **Use Cases:** Data visualization, exploratory analysis.

### 4. Uniform Manifold Approximation and Projection (UMAP)
**Non-linear** technique balancing local and global structure preservation.
- **Key Features:** Scalable, preserves both local and global data structures.
- **Time Consumption:** \(O(nlogn)\) complexity; faster and more scalable than t-SNE.
- **Use Cases:** Visualization, clustering, preprocessing.
- **Practical Advantage:** Superior in both speed and performance compared to other methods, making it ideal for large datasets.

### 5. Autoencoders
**Neural network-based** technique for feature extraction.
- **Key Features:** Captures non-linear relationships, customizable architecture.
- **Time Consumption:** Requires significant training time and computational resources.
- **Use Cases:** Dimensionality reduction, anomaly detection, data denoising.

### 6. Non-negative Matrix Factorization (NMF)
**Linear** technique that factors data into non-negative matrices.
- **Key Features:** Additive parts-based representations, interpretable components.
- **Time Consumption:** Iterative and can be time-consuming, but faster with sparse data.
- **Use Cases:** Text mining, image processing, recommendation systems.


## Comparison of Dimensionality Reduction Methods

| Method        | Type       | Computational Complexity | Scalability        | Best For                             |
|---------------|------------|--------------------------|--------------------|--------------------------------------|
| PCA           | Linear     | \(O(n^3)\)               | Limited for very large datasets | Noise reduction, feature visualization |
| LDA           | Linear     | Similar to PCA           | Limited for very large datasets | Supervised classification tasks     |
| t-SNE         | Non-linear | \(O(n^2)\)               | Poor scalability    | High-quality visualizations          |
| UMAP          | Non-linear | \(O(n \log n)\)          | Highly scalable     | Visualization and clustering of large datasets |
| Autoencoders  | Non-linear | Varies (requires training)| Depends on architecture | Complex feature extraction           |
| NMF           | Linear     | Iterative, varies        | Moderate scalability | Interpretable component extraction   |


### Randomness :
Many non-linear dimension reduction methods incorporate randomness by default, whereas linear methods like PCA are typically deterministic. However, some non-linear methods can be made deterministic by fixing seeds, and certain linear methods can also use randomized approaches.

## Time Consumption and Performance

- **PCA & LDA:** Efficient for smaller datasets but become slow as size grows.
- **t-SNE:** High-quality visuals but \(O(n^2)\) complexity limits large-scale use.
- **UMAP:** Comparable visualization quality to t-SNE with \(O(n \log n)\) complexity, making it faster and more scalable.
- **Autoencoders:** Powerful but require extensive training time and resources.
- **NMF:** Interpretable results with iterative processing; faster on sparse data.

### Practical Advantage of UMAP

UMAP is often the preferred dimensionality reduction method in practice due to its:

- **Time Efficiency:** Faster than t-SNE and scales better with large datasets.
- **Performance:** Preserves both local and global data structures effectively.
- **Flexibility:** Easily integrates into various machine learning pipelines for visualization and preprocessing.

## Applications

- **Visualization:** Plotting high-dimensional data in 2D/3D.
- **Feature Extraction & Selection:** Enhancing model performance with relevant features.
- **Preprocessing for Machine Learning:** Reducing computational costs and speeding up training.
- **Noise Reduction:** Improving model accuracy by eliminating irrelevant information.

## Benefits

- **Simplified Models:** Easier to interpret with fewer features.
- **Enhanced Performance:** Faster training and inference, reduced computational resources.
- **Mitigated Overfitting:** Lower risk of learning noise from the data.
- **Improved Visualization:** Clearer representation of complex data structures.

## When to Use

- **Unsupervised Tasks:** Exploratory analysis, clustering, noise reduction.
- **Supervised Tasks:** Classification, regression, enhancing predictive modeling through feature reduction.

## Conclusion

Dimensionality reduction transforms complex, high-dimensional data into manageable forms, enhancing machine learning models' performance and interpretability. Among various methods, **UMAP** stands out for its excellent balance of computational efficiency and performance, making it a practical choice for large-scale and real-world applications.

# Understanding Manifolds in Geometry and Dimension Reduction

## Types of Geometry

### Euclidean Geometry
- **2D Euclidean Geometry**: Represents the flat, two-dimensional plane we commonly interact with.
- **3D Euclidean Geometry**: Extends into three dimensions, encompassing the space around us.

## Manifolds

- **Definition**: A manifold is a multi-dimensional space that, on a small scale, resembles Euclidean space.
- **Example**:
  - **Earth as a Manifold**: While the Earth is a 3D spherical object globally, any small, localized region appears flat and 2D, aligning with Euclidean geometry. This concept was historically used when early explorers considered the Earth flat due to the limited scope of their observations.

## Manifold Learning in Dimension Reduction

Non-linear dimension reduction methods leverage the concept of manifolds to effectively reduce high-dimensional data into lower dimensions while preserving essential structures.

### UMAP (Uniform Manifold Approximation and Projection)
- **Approach**: Assumes that data lies on a low-dimensional manifold embedded in a higher-dimensional space.
- **Functionality**: Constructs a graphical representation of the manifold and optimizes a low-dimensional projection that maintains the manifold's structure.

### t-SNE (t-Distributed Stochastic Neighbor Embedding)
- **Approach**: Focuses on preserving the local neighborhood structure of data points.
- **Functionality**: Converts high-dimensional distances into probabilities and minimizes the divergence between these probabilities in the low-dimensional space, effectively capturing the manifold's shape.

## Summary

Manifolds provide a foundational framework for understanding complex, high-dimensional data by acknowledging that such data often reside on lower-dimensional, locally Euclidean surfaces. Techniques like UMAP and t-SNE exploit this property to perform effective non-linear dimension reduction, enabling better visualization and analysis of intricate datasets.



# Increasing Feature Space with Dimension Reduction Techniques!

## Introduction

In data science, it's common to encounter large datasets with a relatively low number of features. While having many samples is beneficial for training robust models, a limited feature set can restrict the model's ability to capture complex patterns. One effective strategy to enrich the feature space is by **applying dimension reduction techniques** to generate additional features.

## Increasing Features Through Dimension Reduction

Contrary to the traditional use of dimension reduction for decreasing the number of features, these techniques can also be employed to create new, meaningful features from the existing ones. Here's how you can achieve this:

### Step-by-Step Process

1. **Start with a Dataset**:
   - **Example**: Suppose you have a dataset with **4 features**.

2. **Apply PCA (Principal Component Analysis)**:
   - **Purpose**: PCA transforms the original features into a new set of orthogonal components that capture the maximum variance in the data.
   - **Outcome**: Applying PCA to the 4 original features yields **4 principal components**.

3. **Concatenate the Principal Components**:
   - **Action**: Combine the original 4 features with the 4 principal components.
   - **Result**: You now have **8 features** (4 original + 4 PCA).

4. **Apply UMAP (Uniform Manifold Approximation and Projection)**:
   - **Purpose**: UMAP is a non-linear dimension reduction technique that preserves the local and global structure of the data.
   - **Outcome**: Applying UMAP to the 8-feature dataset generates **8 new UMAP features**.

5. **Concatenate Again**:
   - **Action**: Combine the existing 8 features with the new 8 UMAP features.
   - **Result**: The feature set expands to **16 features** (8 original + 8 UMAP).

6. **Iterate as Needed**:
   - **Flexibility**: You can continue this process, applying other dimension reduction or feature extraction techniques to further enrich the feature space.

## Summary of Unbiased Dimensionality Reduction
Why do we apply scaling, before dimensionality reduction?
- **Definition:**  
  Unbiased dimensionality reduction ensures all features contribute equally when reducing dataset dimensions, preventing any single feature from dominating the process.

- **Importance:**  
  - **Equal Contribution:** Scaling features (e.g., standardization) ensures fair treatment.
  - **True Variance Capture:** Accurately reflects the underlying data structure.
  - **Improved Interpretability:** Results are meaningful and easier to understand.

- **Example with PCA:**  
  - **Without Scaling:** Features with larger scales (e.g., Income vs. Age) may overshadow others.
  - **With Scaling:** All features are standardized, allowing PCA to capture genuine relationships.

- **Benefits:**  
  - Prevents unintended bias from feature scales.
  - Enhances the reliability of downstream machine learning models.

# Cumulative Explained Variance vs Explained Variance vs Explained Variance Ratio in PCA

In Principal Component Analysis (PCA), understanding **explained variance**, **explained variance ratio**, and **cumulative explained variance** is essential for interpreting how much information each principal component (PC) retains from the original dataset.

---

## Explained Variance
- **What it is**: The amount of variance in the data captured by a single principal component (PC).
- **Formula**:
  - Explained Variance of PC_i = Eigenvalue of PC_i
- **Meaning**: Represents the absolute variance explained by a specific PC.

---

## Explained Variance Ratio
- **What it is**: The proportion of the total variance explained by a single PC.
- **Formula**:
  - Explained Variance Ratio of PC_i = (Explained Variance of PC_i) / (Total Variance)
- **Meaning**: Shows the relative importance of a PC in explaining the total variance. It is often expressed as a percentage (e.g., 50%).

---

## Cumulative Explained Variance
- **What it is**: The total proportion of variance explained by the first k principal components.
- **Formula**:
  - Cumulative Explained Variance = Sum of Explained Variance Ratios from PC1 to PCk
- **Why It Works**: Principal components are **perpendicular (orthogonal)** to each other, meaning they are uncorrelated. This ensures that the variance explained by one PC does not overlap with the variance explained by another. As a result, summing their explained variance ratios does not lead to double-counting or errors.
- **Meaning**: Helps determine how much variance is retained when using the first k PCs. It is also expressed as a percentage (e.g., 80%).

---

## Eigenvalues and Their Role in PCA
- **What is an Eigenvalue?**:
  - In PCA, eigenvalues are derived from the covariance matrix of the dataset.
  - Each eigenvalue corresponds to a principal component and represents the **amount of variance** captured by that PC.
  - Larger eigenvalues indicate that the corresponding PC captures more variance in the data.

- **Why Eigenvalue Equals Variance**:
  - PCA works by diagonalizing the covariance matrix of the dataset. The eigenvalues of this matrix represent the variance along the directions of the principal components.
  - Mathematically, the eigenvalue of a PC is equal to the variance of the data when projected onto that PC. This is because the covariance matrix captures the spread (variance) of the data, and its eigenvalues quantify this spread along the principal directions.

---

## Key Differences
| **Metric**                | **Description**                                                                 | **Example**                     |
|---------------------------|---------------------------------------------------------------------------------|---------------------------------|
| **Explained Variance**     | Absolute variance captured by a single PC.                                      | PC1 explains 5 units of variance. |
| **Explained Variance Ratio** | Proportion of total variance explained by a single PC.                        | PC1 explains 50% of the variance. |
| **Cumulative Explained Variance** | Total proportion of variance explained by the first k PCs.           | First 2 PCs explain 80% of the variance. |

---

## Practical Use
- **Explained Variance Ratio**: Helps identify the importance of individual PCs.
- **Cumulative Explained Variance**: Guides the decision on how many PCs to retain for dimensionality reduction (e.g., keep enough PCs to explain 95% of the variance).

---

## Example
Suppose:
- PC1 explains 50% of the variance.
- PC2 explains 30% of the variance.
- PC3 explains 10% of the variance.

Then:
- **Explained Variance Ratio** of PC1 = 50%.
- **Cumulative Explained Variance** after PC1 and PC2 = 50% + 30% = 80%.
- **Cumulative Explained Variance** after PC1, PC2, and PC3 = 50% + 30% + 10% = 90%.

This means the first 3 PCs together explain 90% of the total variance in the dataset. Since PCs are perpendicular (orthogonal), summing their explained variance ratios is valid and does not cause issues.

```python

# Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot cumulative explained variance
# plt.figure(figsize=(8, 5))
plt.scatter(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance vs Number of Principal Components')
plt.grid(True)

# Add a horizontal line at 95% (or any other threshold)
plt.axhline(y=0.95, color='r', linestyle='-', label='95% Explained Variance')
plt.legend()
plt.show('number of values ')

#how to choose the optimal value for number of components
components=np.argmax(cumulative_explained_variance>.90)+1
print('number of components to preserve 90% of informations' ,components)
```
